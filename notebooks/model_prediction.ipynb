```python
import os
import joblib
import numpy as np
```

```python
# -------------------------
# LOAD MODEL + THRESHOLD
# -------------------------
BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))

model = joblib.load(os.path.join(BASE_DIR, "models", "model.pkl"))
threshold = joblib.load(os.path.join(BASE_DIR, "models", "threshold.pkl"))
```

```python
# -------------------------
# PREDICT FUNCTION
# -------------------------
def predict_loan_risk(data: dict):
    """
    Predict loan default risk with probability and risk level.
    """

    feature_order = [
        "age",
        "income",
        "loan_amount",
        "savings_balance",
        "missed_payments",
        "loan_duration",
        "debt_to_income",
        "savings_ratio",
        "payment_stress"
    ]

    input_array = np.array([[data[f] for f in feature_order]])

    probability = model.predict_proba(input_array)[0][1]
    prediction = int(probability > threshold)

    # Risk categorization
    if probability > 0.75:
        risk = "HIGH"
    elif probability > 0.4:
        risk = "MEDIUM"
    else:
        risk = "LOW"

    return {
        "default_prediction": prediction,
        "risk_probability": float(probability),
        "risk_level": risk,
        "threshold_used": float(threshold)
    }
```
